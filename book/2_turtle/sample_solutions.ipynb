## Question 1

```python
import matplotlib.pyplot as plt
import numpy as np

# Function definitions

def start(figsize=(5,5)):
    state[0] = 0
    state[1] = 0
    state[2] = 0
    
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111)
    ax.set_aspect('equal', adjustable='box')

def draw_forward(dis):
    x = state[0]
    y = state[1]
    angle = state[2]
    state[0] = x + np.cos(angle) * dis
    state[1] = y + np.sin(angle) * dis
    plt.plot([x, state[0]], [y, state[1]], color="black", linewidth=2)
    
def move_forward(dis):
    x = state[0]
    y = state[1]
    angle = state[2]
    state[0] = x + np.cos(angle) * dis
    state[1] = y + np.sin(angle) * dis
    
def rotate_left(theta):
    state[2] = state[2] + theta * np.pi / 180
    
state = [0, 0, 0]
```

```python
# Example: draw a unit square
start()
draw_forward(1)
rotate_left(90)
draw_forward(1)
rotate_left(90)
draw_forward(1)
rotate_left(90)
draw_forward(1)
```

```python
# Example: draw a unit hexagon
start()
draw_forward(1)
rotate_left(60)
draw_forward(1)
rotate_left(60)
draw_forward(1)
rotate_left(60)
draw_forward(1)
rotate_left(60)
draw_forward(1)
rotate_left(60)
draw_forward(1)
```

## Question 3

Defining functions

```python
# Function defintion to draw a square
def draw_square():
    for i in range(4):
        draw_forward(1)
        rotate_left(90)
        

start()
draw_square()
```

```python
# Function definitions for other unit polygons
def draw_triangle():
    for i in range(3):
        draw_forward(1)
        rotate_left(120)

def draw_pentagon():
    for i in range(5):
        draw_forward(1)
        rotate_left(72)
        
def draw_hexagon():
    for i in range(6):
        draw_forward(1)
        rotate_left(60)

def draw_heptagon():
    for i in range(7):
        draw_forward(1)
        rotate_left(51.4285714)
```

```python
# Creating a row of polygons

start() # Turtle is placed at (0,0)
draw_triangle() # Turtle finishes back at (0,0)

# Move the turtle forwards to (2,0)
move_forward(2)
draw_square() # Turtle finishes back at (2,0)

# Move the turtle forwards to (4,0)
move_forward(2)
draw_pentagon() # Turtle finishes back at (4,0)

# Move the turtle forwards to (6,0)
move_forward(2)
draw_hexagon() # Turtle finishes back at (6,0)

# Move the turtle forwards to (8,0)
move_forward(2)
draw_heptagon()
```


The following function draws any polygon with `n` sides.
The function has an additional, optional argument `l`, which specifies the side length of the polygon. 
You can tell it is optional because of the `l=...` syntax, which tells the function which value to use for `l` if it isn't defined. Search "Python keyword arguments" if you want to learn more!

In this case, if l isn't defined, it's value defualts to `None`, which is a special empty variable. This triggers a *conditional statement* `if not l`, and the side length defaults to `1/n`.

```python
def draw_ngon(n, l=None):
    # Default value for side length l is 1/n, unless otherwise specified
    if not l:
        l=1/n
    # Repeat for n sides...
    for i in range(n):
        # Draw one side with length l
        draw_forward(l)
        # Rotate by the exterior angle 360˚/n
        rotate_left(360/n)
```

```python
# To create the 10 circle figure, draw m=10 circles, 
# approximated as n=100-sided polygons
n = 100
m = 10

start()

for j in range (m):
    draw_ngon(n, l=(2/n))
    move_forward(1)
    rotate_left(360/m)
```

```python
# Adjusting the values for n, m, and l give different effects
n = 50
m = 15

start()

for i in range(m):
    draw_ngon(n, l=(3/n))
    move_forward(1)
    rotate_left(360/m)
```

## Question 5

```python
# Iteration 1

start(figsize=(10,5))

# This chunk of code draws one spike
draw_forward(1)
rotate_left(60)
draw_forward(1)
rotate_left(240)
draw_forward(1)
rotate_left(60)
draw_forward(1)

plt.title("Iteration 1")
```

Can we create a function that can generalise to any number of iterations, without needing exponential numbers of lines of code? Yes!

Using a slightly mind-bending structure called a *recursive function*. Such a function calls itself, from within itself! It's very important that a recursive function has some sort of exit condition, so that it doesn't create an infinite process that never ends and possibly crashes your computer!

The in the recursive function below, `if n==1` is the exit condition. Once it hits this, we've reached the bottom of the endless rabbit whole of recursion. 
For all other iterations, the value of n is recursively decreased (`koch(n-1)`), so we are guaranteed to eventually reach `n=1`, allowing the recursion to end.

Be careful though, as the number of iterations `n` increases, the number of times the simple spike unit must be drawn scales as $4^{n-1}$! So for 9 iterations, thats $4^8=65536$ recursions!

```python
# Recursive function for a koch fractal
def koch(n):
    
    # Sequence of instructions for a single koch spike unit
    if n == 1:
        draw_forward(1)
        rotate_left(60)
        draw_forward(1)
        rotate_left(240)
        draw_forward(1)
        rotate_left(60)
        draw_forward(1)
    # For iterations above greater than 1, recursively call koch() until n=1
    else:
        koch(n-1)
        rotate_left(60)
        koch(n-1)
        rotate_left(240)
        koch(n-1)
        rotate_left(60)
        koch(n-1)
```

```python
start(figsize=(15,10))

n = 5
koch(n)

plt.title("Iteration {0}".format(n))
```
