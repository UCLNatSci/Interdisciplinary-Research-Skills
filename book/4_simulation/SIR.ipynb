```python
import numpy as np
import matplotlib.pyplot as plt

a = 0.1
b = 0.000006

n_days = 1000
pop_S = np.zeros(n_days + 1)
pop_I = np.zeros(n_days + 1)


pop_S[0] = 20000
pop_I[0] = 100

for i in range(n_days):
    pop_S[i + 1] = pop_S[i] - b * pop_S[i]*pop_I[i]
    pop_I[i + 1] = pop_I[i] + b * pop_S[i]*pop_I[i] - a * pop_I[i]
    
plt.plot(pop_I)
plt.plot(pop_S)
```
